```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\guded\\OneDrive\\Desktop\\INFOSYS\\train-balanced-sarcasm.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.drop('author',axis = 1)
df.head()
```

```python
df = df.drop('subreddit', axis=1)
df.head()
```

```python
df = df.drop('date', axis=1)
df.head()
```

```python
df = df.drop('created_utc', axis=1)
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.count()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
value_counts = df['label'].value_counts().sort_index()
labels = value_counts.index.to_numpy()  
values = value_counts.values
plt.bar(labels, values, color=['skyblue', 'lightgreen'])
plt.xlabel("Label Value")
plt.ylabel("Count")
plt.title("Distribution of 0 and 1 in 'label' Column")
plt.xticks([0, 1], ['0', '1'])
plt.grid(axis='y', linestyle='--', alpha=0.6)
```

```python
plt.tight_layout()
plt.show()
```
